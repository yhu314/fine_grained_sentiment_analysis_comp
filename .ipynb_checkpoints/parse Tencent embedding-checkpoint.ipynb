```python
import numpy as np
import pickle
```

```python
path = '../tencent_embedding/Tencent_AILab_ChineseEmbedding.txt'
```

```python
vocab_path = '../tencent_embedding/vocab.pickle'
array_test_path = '../tencent_embedding/embedding.txt'
```

```python
raw_file = open(path, 'rb')
array_file = open(array_test_path, 'w', encoding='utf8')
```

```python
vocab = dict()

while True:
    line = raw_file.readline()
    if not line:
        break

    try:
        line = line.decode('utf-8').rstrip('\n')
    except:
        # print(line)
        continue

    word, embed_string = line.split(' ', maxsplit=1)
    
    if word == '8825658':
        continue
        
    if len(embed_string.split(' ')) != 200:
        print(word)
        print(embed_string)
        continue
    
    vocab[word] = len(vocab)

    array_file.write(embed_string+'\n')
    if len(vocab) % 100000 == 0:
        print('{} lines parsed'.format(len(vocab)))

raw_file.close()
array_file.close()
```

```python
with open(vocab_path, 'wb') as file:
    pickle.dump(vocab, file, pickle.HIGHEST_PROTOCOL)
```

```python
with open(vocab_path, 'rb') as file:
    vocab = pickle.load(file)
```
